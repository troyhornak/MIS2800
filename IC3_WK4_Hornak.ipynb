```python
%%sql
CREATE TABLE IF NOT EXISTS stock_preference(
  ticker TEXT,
  price_before_game REAL,
  price_week_after_game REAL,
  trading_volume_millions INTEGER
) STRICT;
```

```python
%%sql
-- creating another table named ad_buys
CREATE TABLE IF NOT EXISTS ad_buys (
  ticker TEXT,
  company_name TEXT,
  ad_cost_millions REAL,
  airtime_seconds INTEGER
) STRICT;
```

```python
%%sql
SELECT ticker, price_before_game FROM stock_preference ORDER BY ticker ASC
```
