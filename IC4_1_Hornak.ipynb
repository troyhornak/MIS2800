```python
%%sql
CREATE TABLE products (
    name TEXT NOT NULL PRIMARY KEY,
    price REAL NOT NULL
);
```

```python
%%sql
CREATE TABLE IF NOT EXISTS orders (
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    product_name TEXT NOT NULL,
    quantity INTEGER NOT NULL,
    order_date DATE
);
```

```python
%%sql
INSERT INTO products (name, price) VALUES
    ('Laptop', 1200.00),
    ('Mouse', 25.50),
    ('Keyboard', 75.00),
    ('Monitor', 350.00),
    ('Webcam', 49.99),
    ('Headphones', 150.00);
```

```python
%%sql
/* Update the price of 'Laptop' */ UPDATE products SET price = 1299.99 WHERE name = 'Laptop';
/* Delete 'Webcam' from the table */ DELETE FROM products WHERE name = 'Webcam';
```
